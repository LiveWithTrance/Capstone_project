```python
import pandas as pd
import numpy as np
```

```python
comment_li = pd.read_excel('Data/annotated_translated_li_comment.xlsx')
comment_aito = pd.read_excel('Data/annotated_translated_aito_comment.xlsx')
```

```python
comment_li['translation']=comment_li['translation'].map(lambda x: x.lower())
```

```python
comment_aito['translation']=comment_aito['translation'].map(lambda x: x.lower())
```

```python


# Check whether there is Chinese
comment_li['contains_chinese'] = comment_li['translation'].str.contains('[\u4e00-\u9fff]', regex=True)
# Check whether there is a word 'translation' 
comment_li['contains_translation'] = comment_li['translation'].str.contains('translation')

```

```python
comment_aito['contains_chinese'] = comment_aito['translation'].str.contains('[\u4e00-\u9fff]', regex=True)

comment_aito['contains_translation'] = comment_aito['translation'].str.contains('translation')
```

```python

comment_aito[comment_aito['contains_translation'] == True]
```

```python

comment_li = comment_li[(comment_li['contains_chinese'] == False) | (comment_li['sentiment'].isna()) | (comment_li['contains_translation']== True)]

```

```python
comment_li.to_excel('Data/cleaned_annotated_translated_li_comment.xlsx')
```

```python

comment_aito = comment_aito[(comment_aito['contains_chinese'] == False) | (comment_aito['sentiment'].isna()) | (comment_aito['contains_translation']== True)]

```
